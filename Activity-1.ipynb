```python
from bs4 import BeautifulSoup
```

```python
import pandas as pd
```

```python
import requests
```

```python
URL='https://www.shopclues.com/laptops.html?'
```

```python
response=requests.get(URL)
```

```python
response
```

```python
back_code=BeautifulSoup(response.text)
```

```python
back_code
```

```python
product_name=back_code.select(".prod_name")
```

```python
product_name
```

```python
product_price=back_code.select(".p_price")
```

```python
offer=back_code.select(".prd_discount")
```

```python
data=pd.DataFrame(product_name,columns=['Laptops'])
```

```python
data["product_price"]=product_price
```

```python
data
```

```python
data["offer"]=offer
```

```python
data
```

```python
data.to_csv("Laptops.csv")
```
